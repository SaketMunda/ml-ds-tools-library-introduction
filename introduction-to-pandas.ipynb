### Fill Missing Data

- fillna() 
     - used to fill the default value or set the passed value for the tabs or rows having NaN or null values in the df
     - takes a value as a parameter which will be set in place of empty values

```python
car_sales_missing["Odometer"].mean()
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())

# rather reassinging we can also use inplace=True parameter into the fillna method
# car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
```

```python
car_sales_missing
```

### Drop rows with Missing data

- dropna()

```python
car_sales_dropped = car_sales_missing.dropna()

# it also takes inplace parameter like fillna()
```

```python
car_sales_dropped
```

```python
car_sales_dropped.to_csv("car-sales-dropped.csv")
```

### Reshuffle or Taking random set of data from Data 

#### Important Fact:
        - Data should not be always in the same order, if it is then while the process of learning, the machines will remember the pattern rather learning, so it's better take a portion of data or shuffle the data for learning process.
        - for example, if we have 1M data then we can use 20% of data for training.

- This can be achieved using sample()
- Takes frac as with value as 0.5 for 50%, 1 as 100% or 0.01 for 1%

```python
car_sales.sample(frac = 0.5)
```

```python
car_sales.sample(frac = 1)
```

```python
# Notice above the index are shuffled
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# Once reshuffled, how to reset the index ?
# using reset_index()

car_sales_shuffled.reset_index(inplace=True)
```

```python
car_sales_shuffled
```

```python
# Notice that extra index column is added, if we don't need that,
# pass drop = True
car_sales_shuffled.reset_index(drop=True,inplace=True)
```

```python
car_sales_shuffled
```
