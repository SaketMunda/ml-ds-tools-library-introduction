## 0. An end-to-end Scikit-learn workflow

```python
# 1. Get the data ready

import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
import numpy as np
```

```python
# Separate the feature matrix and target matrix

X = heart_disease.drop("target", axis=1) # feature variables
Y = heart_disease["target"] # target variable or labels
```

```python
# 2. Choose the right model and hyperparameters

from sklearn.ensemble import RandomForestClassifier

# Random Forest Classifier is the machine learning model for classification since we're doing a classfication problem
```

```python
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model/algorithm to the training data

# first, split the data into training and test set
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
clf.fit(X_train, Y_train)
```

```python
# make predictions
Y_label = clf.predict(np.array([1,2,3,4]))
```

```python
# throwing error above because, our model only learnt what we have taught to understand 2-D arrays 
# and it trained with X_train/X_test shape

y_preds = clf.predict(X_test)
y_preds
```

```python
# and note that the above output looks like something Y_test
Y_test
```

```python
# 4. Evaluate the model on training data and test data
clf.score(X_train,Y_train) # .score gives the mean accuracy of the model
```

```python
clf.score(X_test,Y_test)
```

```python
# other metrics

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(Y_test, y_preds))
```

```python
confusion_matrix(Y_test, y_preds)
```

```python
accuracy_score(Y_test, y_preds)
```

```python
# 5. Improve a model

# Try different amount of n_estimators (one of the estimators)

np.random.seed(42) # so everytime take same numbers of seed(40)

for i in range(10,100,10):
    print(f"Trying the model with n_estimators = {i}")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, Y_train)
    print(f"Accuracy :{clf.score(X_test,Y_test) * 100:.2f}%")    
```

```python
# 6. Save a model and load it

import pickle # python library to read and write

pickle.dump(clf, open("random_forest_model_1.pkl", "wb")) # wb : write binary

```

```python
# load model

loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb")) # rb: read binaries
loaded_model.score(X_test, Y_test)
```

## 1. Getting our data ready to be used with Machine Learning

Three main things we have to do:
   1. Split the data into features and target(labels), usually 'X' and 'y'
   2. Filling(also called imputing) or disregarding missing values
   3. Converting non-numerical values to numerical values (also called feature encoding), as machine understands only 0 and 1.

```python
# first we need to import the standard libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import the data or create the data
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# 1. Splitting data into feature variable and target Variable from data imported

X = heart_disease.drop("target", axis = 1)
y = heart_disease["target"]

```

```python
y.head()
```

```python
# Split the data into training and test sets
# according to Machine Learning workflow, we must separate the data for training and evaluation
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

### 1.1 Make sure all the data is numerical

```python
car_sales = pd.read_csv("car-sales-extended.csv")
car_sales

# Problem we are solving here, to predict Price on the basis of given Make, Colour, Odometer, Doors

# This is a regression problem of prediction
```

```python
# Split the data into X and y

X = car_sales.drop("Price", axis = 1)
y = car_sales["Price"]
```

```python
# Split into Train and Test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Build Machine learning model
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### So here we can see that machine can't understand strings, so we have to convert string data into numerical

```python
X.head()
```

```python
# Turn the categories into numbers

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

one_hot = OneHotEncoder()
categorical_feature = ['Make','Colour','Doors']

# Here Doors are also a categorical feature, since some cars have 4 doors, some have 5 and some have 3

transformer = ColumnTransformer([('one_hot', one_hot, categorical_feature)], remainder='passthrough')

transformed_X = transformer.fit_transform(X)
```

```python
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
# Let's refit the model

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)

model.fit(X_train, y_train)

model.score(X_test, y_test)
```
