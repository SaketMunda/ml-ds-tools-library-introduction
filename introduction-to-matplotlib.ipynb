## Matplotlib Workflow

It contains the steps which transforms a *raw data* in form of numbers or spreadsheets into figures, charts and graphs.


Steps involved are,

- Create/Import the data
- Create empty figure/plot
- Plot data (axes on figure)
- Customize plot with some labels and titles
- Save/Share plot


```python
# Step 1 : Create the data
x = [1,2,3,4]
y = [11,2,33,44]

# Step 2 : Create a figure/plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# Step 3: Plot data
ax.plot(x,y)

# Step 4: Customize plot with some titles and labels
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# Step 5: Save plot
fig.savefig("images/sample-plot.png")
```

### Subplots

Subplots can be plotted in two ways,

#### First Method

This approach is in the form of tuples as you can see below that axes are declared in form tuples
- but the disadvantage I personally see is if I want to plot more number of plots in a single figure then it's really time consuming to declare with different names.

```python
fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,10))
# Filling data
ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(prices.keys(), prices.values())
ax3.hist(x)


# Customizing
ax0.set(title="This plot is ax0")
ax1.set(title="This plot is ax1")
ax2.set(title="This plot is ax2")
ax3.set(title="This plot is ax3");

# Title for the figure
fig.suptitle("Subplots Type 1", fontsize=16);
```

#### Second Method

This method is achieved using list of axes, ax[0,0], ax[0,1] something like that

- What I observed that since this is like an 2-d array, so it can be transformed into loop and plot the graphs according to the index.
- no need to declare with multiple names in the form of tuples

```python
fig, ax = plt.subplots(nrows=2, ncols = 2, figsize=(10,10))

# filling data
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.randn(1000));


fig.suptitle("Subplots Type 2", fontsize=20);
```

### We have seen many ways to plot a figure and graphs, using pyplot API and Object Oriented matplotlib API, So which one to use ?

Ideally, it depends on the use of that figure, 
* If you want to plot it quickly with less customization or with default settings just to have a rough visualization then should use pyplot integrated API
* But if want to customize as to make use of all the features of matplotlib plotting then must use Object oriented API for visualization

```python
# selecting records which has age above 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Let's plot using pyplot API
over_50.plot(kind='scatter',x="age",y="chol",c="target", figsize=(10,6), ylabel="Cholestrol", xlabel="Age");
```

```python
# plot using object oriented API mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x= "age",y="chol",c="target", ax=ax);

#ax.set_xlim([45,100]); or plt.xlim([45,100]); #it sets the x limit from 45-100, uncomment to see
```

```python
# plot using object oriented API from scratch

# create figure
fig, ax = plt.subplots(figsize=(10,6))

# add data
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"])

# plot the figure
ax.plot()


# customize the plot
ax.set(title="Heart Disease & Cholestrol Level", xlabel="Age", ylabel="Cholestrol");

ax.legend(*scatter.legend_elements(),title="Target");

ax.axhline(over_50["chol"].mean() ,linestyle="--");
```

```python
# plotting subplots under a single figure

# create a fig
fig, ax = plt.subplots(nrows=2,ncols=1 ,figsize=(10,10), sharex = True);


# fill data

scatter0 = ax[0].scatter(over_50["age"], over_50["chol"], c=over_50["target"])
scatter1 = ax[1].scatter(over_50["age"], over_50["thalach"], c=over_50["target"]);

# customize plot
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")
ax[0].set(title="Heart Disease & Cholestrol Levels", ylabel="Cholestrol")
ax[1].set(title="Heart Disease & Max Heart Rate", xlabel="Age", ylabel="Heart Rate");

# add legends
ax[0].legend(*scatter0.legend_elements(), title="Target")
ax[1].legend(*scatter1.legend_elements(), title="Target")

# add meanline
ax[0].axhline(over_50["chol"].mean(), linestyle="--")
ax[1].axhline(over_50["thalach"].mean(), linestyle="--");
```
