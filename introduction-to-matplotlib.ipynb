There are several approaches of ploting the figures using matplotlib

- pyplot API
- object-oriented API

Generally, the pyplot API is less flexible than the object-oriented API and also **Recommended**

```python
x=[1,2,3,4]
y=[11,22,33,44]

# first method

fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes

plt.show()
```

```python
# 3rd method (Recommended)

fig, ax = plt.subplots() # creates figure

ax.plot(x,[50,100,200,150]);

ax.plot(x,[80,120,220,250]);
```

## Matplotlib Workflow

It contains the steps which transforms a *raw data* in form of numbers or spreadsheets into figures, charts and graphs.


Steps involved are,

- Create/Import the data
- Create empty figure/plot
- Plot data (axes on figure)
- Customize plot with some labels and titles
- Save/Share plot


```python
# Step 1 : Create the data
x = [1,2,3,4]
y = [11,2,33,44]

# Step 2 : Create a figure/plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# Step 3: Plot data
ax.plot(x,y)

# Step 4: Customize plot with some titles and labels
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# Step 5: Save plot
fig.savefig("images/sample-plot.png")
```

### bar plots

**Use case for bar plots are when plotted using Dictionary**

For example,
- Using Nut butter price

```python
prices = {"Almond Butter":10,"Peanut Butter":8,"Cashew Butter": 20}

fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values()); # height is y value
ax.set(title="Nut Butter Store", ylabel="Price ($)");
```

```python
# There is one more way of Bar plot representation,
# Horizontal Bar Plotting

fig, ax = plt.subplots()
ax.barh(list(prices.keys()), list(prices.values()));
```

### Subplots

Subplots can be plotted in two ways,

#### First Method

This approach is in the form of tuples as you can see below that axes are declared in form tuples
- but the disadvantage I personally see is if I want to plot more number of plots in a single figure then it's really time consuming to declare with different names.

```python
fig, ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,10))
# Filling data
ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(prices.keys(), prices.values())
ax3.hist(x)


# Customizing
ax0.set(title="This plot is ax0")
ax1.set(title="This plot is ax1")
ax2.set(title="This plot is ax2")
ax3.set(title="This plot is ax3");

# Title for the figure
fig.suptitle("Subplots Type 1", fontsize=16);
```

#### Second Method

This method is achieved using list of axes, ax[0,0], ax[0,1] something like that

- What I observed that since this is like an 2-d array, so it can be transformed into loop and plot the graphs according to the index.
- no need to declare with multiple names in the form of tuples

```python
fig, ax = plt.subplots(nrows=2, ncols = 2, figsize=(10,10))

# filling data
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.randn(1000));


fig.suptitle("Subplots Type 2", fontsize=20);
```

## Plotting using Pandas DataFrames

**Pandas** have a rich feature to visualise data internally using **Matplotlib**, so let's try plotting with pandas.

```python
# create/import a dataframe

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales.plot("Odometer (KM)","Doors");

# Plotting anything doesn;t mean it's correct, as a datascience engineer you have to plot meaningful data 
```

```python
# Lets plot some meaningful data


# This kind of figure we will try to plot using car_sales dataframe
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2022", periods=1000))

ts = ts.cumsum() # for cumulative sum

# for example, we have dataframe as 1,2,3,4,5
# then cumsum() would be, 1,3,6,10,15

# ploting directly
ts.plot();
```

```python
# let's do with our car_sales data

# first need to add some column like Date

car_sales["Sale Date"] = pd.date_range("1/1/2022", periods=len(car_sales)) 
car_sales
```

```python
# suppose these cars are sold by a company called, Cars24, THEN what would be the cumulative sales after each sale
# date

car_sales["Price"] = car_sales["Price"].str.replace('[\$\,]','').astype(float).astype(int) 
car_sales["Total Sales ($)"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot("Sale Date", "Total Sales ($)");
```

```python
car_sales.plot("Sale Date", "Total Sales ($)", kind='pie');
```
