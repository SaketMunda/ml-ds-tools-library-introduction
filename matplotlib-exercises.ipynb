First we'll need to import pandas and create a DataFrame work with.

```python
# Import pandas as pd
import pandas as pd
```

```python
# Import the car-sales.csv into a dataframe called car_sales and view
car_sales = pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
# try to plot the Price column using plot() column
plt.plot(car_sales["Price"]);
```

This is not something meaningful, as it's not numeric data that we are trying to plot.

So first we need convert it to numerical.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales["Price ($)"] = car_sales["Price"].str.replace("[\$\,]","").astype(float).astype(int)
```

```python
car_sales.drop("Price", axis=1, inplace = True)
```

```python
# Add a column called 'Total sales' to car_sales which cumulatively adds the 'Price' column
car_sales["Total Sales ($)"] = car_sales["Price ($)"].cumsum()

#Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales["Sale Date"] = pd.date_range(start="1/1/2022",periods=len(car_sales))


# View the car_sales DataFrame
car_sales.head()
```

```python
car_sales.head()
```

Now we have got a numeric column `Total Sales` and a dates column `Sale Date`, let's visualize them

```python
# use plot() function to plot the `Sale Date` column versus the `Total Sales` column
plt.plot(car_sales["Sale Date"], car_sales["Total Sales ($)"]);
```

```python
# Create a scatter plot of the `Odometer (KM)` and `Price` column using the plot() function
plt.scatter(car_sales["Odometer (KM)"], car_sales["Price ($)"]);
```

```python
# Create a Numpy array of random numbers of size (10,4) and save it to X
X = np.random.randint(50,size=(10,4))

# Turn the Numpy array into a Dataframe with columns called [a,b,c,d]
df = pd.DataFrame(X,columns=['a','b','c','d'])

# CReate a bar graph of the dataframe\
df.plot(kind="bar")
```

```python
# create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame

plt.bar(car_sales["Make"], car_sales["Odometer (KM)"]);
```

```python
# create a histogram of the 'Odometer (KM)' column
plt.hist(car_sales["Odometer (KM)"]);
```

```python
# Create a histogram of the 'Price' column with 20bins
plt.hist(car_sales["Price ($)"], bins=20);
```

Let's try different dataset

```python
# Import heart_disease.csv and save it to the variable 'heart_disease'

heart_disease = pd.read_csv("heart-disease.csv")
```

```python
# View the first 10 rows of the heart_disease DataFrame
heart_disease[:10]
```

```python
# Create a histogram of the "age" column with 50 bins
plt.hist(heart_disease.age, bins=50);
```

```python
# call plot.hist() on the heart_disease DataFrame and toggle the "subplots" parameter to True
heart_disease.plot.hist(subplots=True);
```

That plot looks pretty squished. Let's change the figsize.

```python
# Call the same line of code from above except change the "figsize" parameter to be (10,30)
heart_disease.plot.hist(figsize=(10,30), subplots= True);
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in patients over 50 years old.

```python
heart_disease.head()
```

```python
# Create a dataframe with patients over 50 years old
heart_disease_above_50 = heart_disease[heart_disease["age"] > 50]
```

```python
heart_disease_above_50
```

```python
# CReate the plot
fig, ax = plt.subplots(figsize=(10,5))

# plot the data
scatter = ax.scatter(heart_disease_above_50["age"], heart_disease_above_50["chol"], c = heart_disease_above_50["target"])


# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax.axhline(heart_disease_above_50["chol"].mean(), linestyle="--");
```

Beautiful, now you've created a plot of two different variables, let's change the style

```python
# Check what styles are available under plt
plt.style.available
```

```python
# Change the style to use "seaborn-whitegrid"
plt.style.use('seaborn-whitegrid')
```

```python
# Reproduce the same figure as above with the "seaborn-whitegrid" style

# Create the plot
fig, ax = plt.subplots(figsize=(10,5))

# plot the data
scatter = ax.scatter(x=heart_disease_above_50["age"], y = heart_disease_above_50["chol"], c = heart_disease_above_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax.axhline(heart_disease_above_50["chol"].mean(), linestyle="--");
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good color.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

```python
# Replot the same figure as above except change the "cmap" parameter of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
sc = ax.scatter(x=heart_disease_above_50["age"], y=heart_disease_above_50["chol"], c=heart_disease_above_50["target"], cmap="winter")

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*sc.legend_elements(), title="Target")

# Add a meanline
ax.axhline(heart_disease_above_50["chol"].mean(),linestyle="--",color="red");
```

Beautiful ! Now our figure has an upgraded color scheme let's save it to file.

```python
# Save the current figure using savefig(), the file name can be anything you want
fig.savefig("heart_disease_age_vs_chol_fig.png")
```
