```python
# Import and get matplotlib ready
import matplotlib.pyplot as plt
%matplotlib inline

# Prepare data( create two lists of numbers, X, y)
X, y = np.random.randint(10, size = 5), np.random.randint(10, size = 5)

# Setup figure and axes using plt.subplots()
fig, ax = plt.subplots()

# Add data (X,y) to axes
ax.plot(X,y)

# Customize plot by adding a title, xlabel, ylabel
plt.title("Matplotlib Workflow")
plt.xlabel("X-axis")
plt.ylabel("y-axis")

# Show the plot
#plt.show()

#Save the plot to file using fig.savefig()
fig.savefig("matplotlib-sample-workflow-exercise-plot.png")
```

First we'll need to import pandas and create a DataFrame work with.

```python
# Import pandas as pd
import pandas as pd
```

```python
# Import the car-sales.csv into a dataframe called car_sales and view
car_sales = pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
# try to plot the Price column using plot() column
plt.plot(car_sales["Price"]);
```

This is not something meaningful, as it's not numeric data that we are trying to plot.

So first we need convert it to numerical.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales["Price ($)"] = car_sales["Price"].str.replace("[\$\,]","").astype(float).astype(int)
```

```python
car_sales.drop("Price", axis=1, inplace = True)
```

```python
# Add a column called 'Total sales' to car_sales which cumulatively adds the 'Price' column
car_sales["Total Sales ($)"] = car_sales["Price ($)"].cumsum()

#Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales["Sale Date"] = pd.date_range(start="1/1/2022",periods=len(car_sales))


# View the car_sales DataFrame
car_sales.head()
```

```python
car_sales.head()
```

Now we have got a numeric column `Total Sales` and a dates column `Sale Date`, let's visualize them

```python
# use plot() function to plot the `Sale Date` column versus the `Total Sales` column
plt.plot(car_sales["Sale Date"], car_sales["Total Sales ($)"]);
```

```python
# Create a scatter plot of the `Odometer (KM)` and `Price` column using the plot() function
plt.scatter(car_sales["Odometer (KM)"], car_sales["Price ($)"]);
```

```python
# Create a Numpy array of random numbers of size (10,4) and save it to X
X = np.random.randint(50,size=(10,4))

# Turn the Numpy array into a Dataframe with columns called [a,b,c,d]
df = pd.DataFrame(X,columns=['a','b','c','d'])

# CReate a bar graph of the dataframe\
df.plot(kind="bar")
```

```python
# create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame

plt.bar(car_sales["Make"], car_sales["Odometer (KM)"]);
```

```python
# create a histogram of the 'Odometer (KM)' column
plt.hist(car_sales["Odometer (KM)"]);
```
